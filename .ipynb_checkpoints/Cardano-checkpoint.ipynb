### LSTM Networks

Long Short Term Memory networks – usually just called “LSTMs” – are a special kind of RNN and  are explicitly designed to avoid the long-term dependency problem. They are very powerful in sequence prediction problems and this is important in our case because the previous price of a stock is crucial in predicting its future price.

#### Creating Data with Timesteps
 
 We start by creating data in  timesteps and converting it into an array using NumPy. Next, we convert the data into a 3D dimension array with X instances, 30 timestamps, and one feature at each step.

```python
# first we select the features for our model, we are going to keep the Close feature
from sklearn.model_selection import train_test_split

ada=ada['Close']
ada=pd.to_numeric(ada,errors='coerce') #turning the Close column to numeric
ada_train, ada_test= train_test_split(ada, train_size=0.80, test_size=0.20,  shuffle=False)

Xtrain=ada_train.values
Xtest=ada_test.values
```

```python
#function to create the input 3D sequence for our network
def create_3d_input(sequence, timestamps):
    
    X = []
    y = []
    
    for i in range(timestamps, len(sequence)):
        X.append(sequence[(i-timestamps):i])
        y.append(sequence[i])
    
    X, y = np.array(X), np.array(y)
    X = np.reshape(X, (X.shape[0], X.shape[1], 1))
    return X,y
```

```python
X_train,y_train=create_3d_input(Xtrain,30)
X_test,y_test=create_3d_input(Xtest,30)
```

```python
#!pip install tensorflow
#!pip install keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
```

Now we are going to create a function to  buil our neural network. We are going to use four internal layers and one output layer. Also we are going to use the Adam optimizer and the Mean Squared Error as loss function.

```python
def build_fit_model():
    regressor = Sequential()
    unit=100
    ## input_shape=timesteps, features
    regressor.add(LSTM(units = unit, return_sequences = True, input_shape = (X_train.shape[1], 1)))
    regressor.add(Dropout(0.2))

    regressor.add(LSTM(units = unit, return_sequences = True))
    regressor.add(Dropout(0.2))

    regressor.add(LSTM(units = unit, return_sequences = True))
    regressor.add(Dropout(0.2))

    regressor.add(LSTM(units = unit))
    regressor.add(Dropout(0.2))

    regressor.add(Dense(units = 1))

    regressor.compile(optimizer = 'adam', loss = 'mean_squared_error')

    regressor.fit(X_train, y_train, epochs = 20, batch_size = 32)
    return regressor
```

As we know the effectiveness of our trained neural network will depends on the weights that we used in our building  process. Therefore, since we want that our results can be reproducible we are going to save the trained model with the best performance by using as metric the R^2.

```python
from sklearn.metrics import r2_score
r2=0
while(r2<0.8):
    regressor=build_fit_model()
    regressor.save_weights('my_weights.model')
    regressor.save('my_model.hdf5')
    predicted_stock_price = regressor.predict(X_test)
    r2=r2_score(y_test[30:],predicted_stock_price[30:])
```

Then we can load our saved model with the highest R^2 and make the predictions. As we can see despite our prediction its a little right shifted it gets the trends very well and we obtain really good R^2.

```python
from keras.models import load_model

model = load_model('my_model.hdf5')
y_pred=model.predict(X_test)
```

```python
plt.plot(y_test, color = 'blue', label = 'Real ADA Stock Price')
plt.plot(y_pred, color = 'green', label = 'Predicted ADA Stock Price')
plt.title('ADA Stock Price Prediction')
plt.xlabel('Time')
plt.ylabel('ADA Stock Price')
plt.legend()
plt.show()
```
